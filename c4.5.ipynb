```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_recall_fscore_support
from sklearn.metrics import roc_auc_score, roc_curve
import time
import seaborn as sn
```

```
X_test = pd.read_csv('Dataset/X_test.csv')
y_test = pd.read_csv('Dataset/y_test.csv').iloc[:, 0]

X_train_30 = pd.read_csv('Dataset/X_train_30.csv')
y_train_30 = pd.read_csv('Dataset/y_train_30.csv').iloc[:, 0]

X_train_60 = pd.read_csv('Dataset/X_train_60.csv')
y_train_60 = pd.read_csv('Dataset/y_train_60.csv').iloc[:, 0]

X_train_100 = pd.read_csv('Dataset/X_train_100.csv')
y_train_100 = pd.read_csv('Dataset/y_train_100.csv').iloc[:, 0]
```

### Train on 30%

```
start_time = time.time()
```

```
c45_30 = DecisionTreeClassifier(criterion='gini', random_state=42)
```

```
c45_30.fit(X_train_30, y_train_30)
```

```
train_time_30 = time.time() - start_time
```

```
start_time = time.time()
```

```
y_pred_30 = c45_30.predict(X_test)
```

```
test_time_30 = time.time() - start_time
```

```
accuracy_30 = accuracy_score(y_test, y_pred_30)
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred_30, average='weighted')
```

```
results = {}
```

```
results['30%'] = {
    'accuracy': accuracy_30,
    'precision': precision,
    'recall': recall,
    'f1': f1,
    'train_time': train_time_30,
    'test_time': test_time_30,
    'train_samples': len(X_train_30)
}
```

```
print(f"Accuracy: {accuracy_30:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"Training time: {train_time_30:.4f} seconds")
print(f"Testing time: {test_time_30:.4f} seconds")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_30))
```

```
# Plot accuracy comparison
plt.figure(figsize=(12, 8))

# Subplot 1: Accuracy
plt.subplot(2, 2, 1)
training_sizes = ['30%', '60%', '100%']
accuracies = [results[size]['accuracy'] for size in training_sizes]
plt.bar(training_sizes, accuracies, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('C4.5 Accuracy vs Training Size')
plt.ylabel('Accuracy')
plt.ylim(0.9, 1.0)

# Subplot 2: Training Time
plt.subplot(2, 2, 2)
train_times = [results[size]['train_time'] for size in training_sizes]
plt.bar(training_sizes, train_times, color=['orange', 'purple', 'brown'])
plt.title('C4.5 Training Time vs Training Size')
plt.ylabel('Training Time (seconds)')

# Subplot 3: F1 Score
plt.subplot(2, 2, 3)
f1_scores = [results[size]['f1'] for size in training_sizes]
plt.bar(training_sizes, f1_scores, color=['gold', 'gray', '#CD7F32'])
plt.title('C4.5 F1 Score vs Training Size')
plt.ylabel('F1 Score')
plt.ylim(0.9, 1.0)

# Subplot 4: Sample Count
plt.subplot(2, 2, 4)
sample_counts = [results[size]['train_samples'] for size in training_sizes]
plt.bar(training_sizes, sample_counts, color=['red', 'green', 'blue'])
plt.title('Training Sample Count')
plt.ylabel('Number of Samples')

plt.tight_layout()
plt.show()
```

```
y_pred_proba_30 = c45_30.predict_proba(X_test)
y_pred_proba_60 = c45_60.predict_proba(X_test)
y_pred_proba_100 = c45_100.predict_proba(X_test)
n_classes = len(np.unique(y_test))
print(f"Number of classes: {n_classes}")
if n_classes == 2:
    roc_auc_30 = roc_auc_score(y_test, y_pred_proba_30[:, 1])
    fpr_30, tpr_30, _ = roc_curve(y_test, y_pred_proba_30[:, 1])
else:

    roc_auc_30 = roc_auc_score(y_test, y_pred_proba_30, multi_class='ovr', average='macro')

print("ROC AUC for 30% training data:", roc_auc_30)
if n_classes == 2:

    roc_auc_60 = roc_auc_score(y_test, y_pred_proba_60[:, 1])
    fpr_60, tpr_60, _ = roc_curve(y_test, y_pred_proba_60[:, 1])
else:

    roc_auc_60 = roc_auc_score(y_test, y_pred_proba_60, multi_class='ovr', average='macro')

print("ROC AUC for 60% training data:", roc_auc_60)


if n_classes == 2:

    roc_auc_100 = roc_auc_score(y_test, y_pred_proba_100[:, 1])
    fpr_100, tpr_100, _ = roc_curve(y_test, y_pred_proba_100[:, 1])
else:

    roc_auc_100 = roc_auc_score(y_test, y_pred_proba_100, multi_class='ovr', average='macro')

print("ROC AUC for 100% training data:", roc_auc_100)

# Summary
print("\n=== ROC AUC SUMMARY ===")
print(f"30% Training Data: {roc_auc_30:.4f}")
print(f"60% Training Data: {roc_auc_60:.4f}")
print(f"100% Training Data: {roc_auc_100:.4f}")
results['30%']['roc_auc'] = roc_auc_30
results['60%']['roc_auc'] = roc_auc_60
results['100%']['roc_auc'] = roc_auc_100
```
