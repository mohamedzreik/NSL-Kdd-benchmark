```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_recall_fscore_support
import time
import seaborn as sn
```

```python
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv').iloc[:, 0]

X_train_30 = pd.read_csv('X_train_30.csv')
y_train_30 = pd.read_csv('y_train_30.csv').iloc[:, 0]

X_train_60 = pd.read_csv('X_train_60.csv')
y_train_60 = pd.read_csv('y_train_60.csv').iloc[:, 0]

X_train_100 = pd.read_csv('X_train_100.csv')
y_train_100 = pd.read_csv('y_train_100.csv').iloc[:, 0]
```

### Train on 30%

```python
start_time = time.time()
```

```python
cart_30 = DecisionTreeClassifier(random_state=42)
```

```python
cart_30.fit(X_train_30, y_train_30)
```

```python
train_time_30 = time.time() - start_time
```

```python
start_time = time.time()
```

```python
y_pred_30 = cart_30.predict(X_test)
```

```python
test_time_30 = time.time() - start_time
```

```python
accuracy_30 = accuracy_score(y_test, y_pred_30)
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred_30, average='weighted')
```

```python
results= {}
```

```python
results['30%'] = {
    'accuracy': accuracy_30,
    'precision': precision,
    'recall': recall,
    'f1': f1,
    'train_time': train_time_30,
    'test_time': test_time_30,
    'train_samples': len(X_train_30)
}
```

```python
print(f"Accuracy: {accuracy_30:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"Training time: {train_time_30:.4f} seconds")
print(f"Testing time: {test_time_30:.4f} seconds")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_30))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import numpy as np

# Fix variable naming - use the actual model objects, not confusion matrices
model_30 = DecisionTreeClassifier(random_state=42)
model_30.fit(X_train_30, y_train_30)

model_60 = DecisionTreeClassifier(random_state=42)
model_60.fit(X_train_60, y_train_60)

model_100 = DecisionTreeClassifier(random_state=42)
model_100.fit(X_train_100, y_train_100)

# Get prediction probabilities
y_pred_proba_30 = model_30.predict_proba(X_test)
y_pred_proba_60 = model_60.predict_proba(X_test)
y_pred_proba_100 = model_100.predict_proba(X_test)

# Check number of classes
n_classes = len(np.unique(y_test))
print(f"Number of classes: {n_classes}")
print(f"Classes: {np.unique(y_test)}")

# Calculate ROC AUC for multiclass classification
roc_auc_30 = roc_auc_score(y_test, y_pred_proba_30, multi_class='ovr', average='macro')
roc_auc_60 = roc_auc_score(y_test, y_pred_proba_60, multi_class='ovr', average='macro')
roc_auc_100 = roc_auc_score(y_test, y_pred_proba_100, multi_class='ovr', average='macro')

print("\n=== ROC AUC Scores ===")
print(f"ROC AUC for 30% training data:  {roc_auc_30:.4f}")
print(f"ROC AUC for 60% training data:  {roc_auc_60:.4f}")
print(f"ROC AUC for 100% training data: {roc_auc_100:.4f}")

# Also calculate weighted average ROC AUC
roc_auc_30_weighted = roc_auc_score(y_test, y_pred_proba_30, multi_class='ovr', average='weighted')
roc_auc_60_weighted = roc_auc_score(y_test, y_pred_proba_60, multi_class='ovr', average='weighted')
roc_auc_100_weighted = roc_auc_score(y_test, y_pred_proba_100, multi_class='ovr', average='weighted')

print("\n=== ROC AUC Scores (Weighted) ===")
print(f"ROC AUC for 30% training data:  {roc_auc_30_weighted:.4f}")
print(f"ROC AUC for 60% training data:  {roc_auc_60_weighted:.4f}")
print(f"ROC AUC for 100% training data: {roc_auc_100_weighted:.4f}")

# Update results dictionary with ROC AUC scores
results['30%']['roc_auc_macro'] = roc_auc_30
results['30%']['roc_auc_weighted'] = roc_auc_30_weighted

results['60%']['roc_auc_macro'] = roc_auc_60
results['60%']['roc_auc_weighted'] = roc_auc_60_weighted

results['100%']['roc_auc_macro'] = roc_auc_100
results['100%']['roc_auc_weighted'] = roc_auc_100_weighted

# Create a summary comparison
print("\n=== Performance Comparison Summary ===")
training_sizes = ['30%', '60%', '100%']
metrics = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc_macro', 'roc_auc_weighted', 'train_time', 'test_time']

for metric in metrics:
    print(f"\n{metric.upper()}:")
    for size in training_sizes:
        if metric in results[size]:
            if 'time' in metric:
                print(f"  {size}: {results[size][metric]:.4f} seconds")
            else:
                print(f"  {size}: {results[size][metric]:.4f}")

# Visualize ROC AUC comparison
plt.figure(figsize=(10, 6))
training_percentages = [30, 60, 100]
macro_scores = [roc_auc_30, roc_auc_60, roc_auc_100]
weighted_scores = [roc_auc_30_weighted, roc_auc_60_weighted, roc_auc_100_weighted]

plt.plot(training_percentages, macro_scores, 'b-o', label='ROC AUC (Macro)', linewidth=2, markersize=8)
plt.plot(training_percentages, weighted_scores, 'r-s', label='ROC AUC (Weighted)', linewidth=2, markersize=8)

plt.xlabel('Training Data Percentage (%)')
plt.ylabel('ROC AUC Score')
plt.title('ROC AUC Performance vs Training Data Size')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(training_percentages)
plt.ylim(0.8, 1.0)

for i, (macro, weighted) in enumerate(zip(macro_scores, weighted_scores)):
    plt.annotate(f'{macro:.3f}', (training_percentages[i], macro),
                textcoords="offset points", xytext=(0,10), ha='center')
    plt.annotate(f'{weighted:.3f}', (training_percentages[i], weighted),
                textcoords="offset points", xytext=(0,-15), ha='center')

plt.tight_layout()
plt.show()

print(f"\n=== Key Insights ===")
print(f"â€¢ Best ROC AUC (Macro): {max(macro_scores):.4f} with {training_percentages[np.argmax(macro_scores)]}% training data")
print(f"â€¢ Best ROC AUC (Weighted): {max(weighted_scores):.4f} with {training_percentages[np.argmax(weighted_scores)]}% training data")
print(f"â€¢ ROC AUC improvement from 30% to 100%: {((roc_auc_100 - roc_auc_30) / roc_auc_30 * 100):.2f}%")
```
